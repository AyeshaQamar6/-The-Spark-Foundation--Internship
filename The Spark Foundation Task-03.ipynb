```python
 sns.catplot(x='Category',y='Profit',data=df,col="Region")
```

```python
plt.figure(figsize=(10,10))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

 Here we can see that the sales of furniture is good but still not profitable.

```python
plt.figure(figsize=(24,12))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show
```

```python
plt.figure(figsize=(24,12))
df.groupby('Sub-Category')['Profit'].agg(['sum']).plot.bar()
plt.show
```

This graph shows that sales of tables is high but producing loss, a neglible loss from bookcases and suppliers, almost negligible sale and profit from fasteners, copiers has the highest profit.

```python
sns.barplot(x=df.Discount, y=df.Profit)
```

```python
sns.barplot(x=df.Region, y=df.Sales)
```

```python
sns.barplot(x=df.Region, y=df.Profit)
```

```python
 sns.barplot(x=df.Region, y=df.Profit, hue=df.Category)
```

The furniture category here is making most lost for central region while technology has highest profit in all regions.

```python
 sns.barplot(x=df.Region, y=df.Profit, hue=df.Segment)
```

In this graph, the West region is producing the highest net profit, East region producing very high profit in home office segment and the lowest profit was in Cental region in consumer segment.

```python
dataplot = df.groupby(['State'])['Sales','Profit'].sum()
dataplot.plot.bar(figsize=(20,10))
```

Showing  some states having loss

```python
dataplot = df.groupby(['State'])['Profit'].sum()
dataplot.plot.bar(figsize=(20,10),color='black')
```

Arizona, Colardo, Florida, Illinois, North Carolina, Ohio, Pennsylvania, Tennesee, Texas are producing loss.
States like Texas, Ohio, Illinois and Pennsylvania are incurring huge loss even when the sale is good.
New York and California are making the highest profit.

```python
dataplot = df.groupby(['State']).Discount.sum()
plt.xlabel('States')
plt.ylabel('Discount')
dataplot.plot.bar(figsize=(20,10),color='purple')
```

Texas is giving very high discounts and thus incurring huge loss. The same is the case with Illinois, Ohio and Pennsylvania.

## Final Conclusion:
##### 1- The sale of tables should be stopped as it is producing high loss.
##### 2- Bookcasses and suppliers are producing negligible loss so their price should be increased or sale should be stopped.
##### 3- For the central region we should definitely look into the furniture category as we are suffering most losses there.
